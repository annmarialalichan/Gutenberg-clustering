```
#defining some functions required to preprocess the data
import sys
import os
import re 
import nltk
# list of stop words 
nltk.download("stopwords", quiet = True) 
from nltk.corpus import stopwords # imports the stopwords from the above code
all_stopwords = stopwords.words("english")
from nltk.stem.porter import PorterStemmer 
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet',quiet=True)

def lower_str(info):                                                                                             
  info = info.lower()
  return info

def removing_punctuations(info):
    info = re.sub("[^a-zA-Z]", " ", info)
    # replacing punctuations with spaces
    info = info.strip()                                                                                        
    return info

def removing_stopwords(info):    
    info = info.split()
    info = [word for word in info if not word in set(all_stopwords)]
    line = ' '.join(info)                                                                                         
    return line


def lemmatization(info):
  info=info.split()
  lemmatizer = WordNetLemmatizer()
  info = [lemmatizer.lemmatize(word) for word in info if not word in set(all_stopwords)]
  line=" ".join(info)
  return line

def stemming(info): 
  info=info.split()
  stemmer = PorterStemmer()
  info = [stemmer.stem(word) for word in info if not word in set(all_stopwords)]
  line=" ".join(info)
  return line
  

def removing_singularwords(info):
  info = info.split()
  info = [word for word in info if len(word)>1]
  line=" ".join(info)
  return line
```

Calculating TF IDF Score for each word in a file.








TF=count of a word in a file d/ number of words in d

IDF=log(total words in the dataset/ occurence of word in D)

TF-IDF= TF*IDF

where, d= a particular book or document
       D= set of all books or document
       TF=term frequency/word frequency
       IDF=inverse document frequency

```
words = cleaned_data.flatMapValues(lambda line: line.split())
Words=words.collect()
```

```
Words
```

```
wordCounts = words.map(lambda word:(word,1)).reduceByKey(lambda a, b: a + b)
wordCounts.collect()
```

```
#calculating TF value
tf=wordCounts.map(lambda a: (a[0][1],(a[0][0],a[1])))
```

```
tf.collect()
```

```
word_present_count=wordCounts.map(lambda a:(a[0][1],(a[0][0],a[1],1)))
```

```
word_present_count.collect()
```

```
word_present=word_present_count.map(lambda a: (a[0],a[1][2])).reduceByKey(lambda a,b: a+b)
```

```
word_present.collect()
```

```
N=wordCounts.map(lambda word: (" ",1)).reduceByKey(lambda a,b: a+b)
n=N.collect()
n
```

```
#calculating IDF value
import math
from pyspark.sql.functions import *
idf=word_present.map(lambda a: (a[0],math.log(n[0][1]/a[1])))
```

```
idf.take(5)
```

```
#calculating TF-IDF value
tf_idf=tf.join(idf).map(lambda a: (a[1][0][0],(a[0],a[1][0][1]*a[1][1]))).sortByKey()
tf_idf.collect()
```

```
#list of all tags and book names
tags=wordCounts.map(lambda x: (x[0][1],1)).reduceByKey(lambda a,b: a+b).sortByKey().map(lambda x: x[0])
tags=tags.collect()
booknames=wordCounts.map(lambda x: (x[0][0],1)).reduceByKey(lambda a,b: a+b).sortByKey().map(lambda x: x[0])
booknames=booknames.collect()
booknames
```

```
len(tags)
```

```
#TF-IDF represented in dataframe format 
#spark.conf.set('spark.sql.pivotMaxValues', u'50000')                           
df=tf_idf.map(lambda a: (a[0],a[1][0],a[1][1]))
df.toDF(["Book Name","Tags","TF-IDF SCORE"]).show()
df = df.collect()
```

```
type(df)
```

```
#flattening into an array
data={}
for i in range(len(df)):
  data[df[i][0],df[i][1]]=df[i][2]
  #print(data)
  #print(i)
data
```

```
#converting into a pandas dataframe
import pandas as pd
data1=pd.DataFrame(index=booknames,columns=tags)
for k,v in data.items():
  data1.loc[k[0],k[1]]=v
data1
```

```
#NaN can be seen since all the words cannot be present in all of these books. Here we have replaced NaN values with 0 being the least
data1.fillna(0,inplace=True)
```

```
data1
```

```
#Clustering the selected documents using feature tags
from sklearn.cluster import KMeans
model=KMeans(n_clusters=8,random_state=0)
model.fit(data1)
model.labels_
```

It can be observed that there are 8 clusters and each of the book can be clustered into any of these 0-7 clusters. The books belonging to the same clusters are similar in content and hence can be kept together.

Program 2: **Searching for books based on a particular author**

```
file='/content/Gutenberg/txt'
```

```
import os
```

```
all_files= os.listdir(file)
```

```
#creating dictionary with Author name as it's key and Book name as it's values
auth_name={}
for i in all_files:
    if not i.startswith("."):
        separator= i.split("___")
        bookname=separator[1]
        author = separator[0].lower()
        if author not in auth_name.keys():
            new = {author: [bookname]}
            auth_name.update(new)
        else:
            auth_name.get(author).append(bookname)
```

```
auth_name
```

```
#taking input from the user and displaying the result
search_author = input("Provide author name: ").lower()
if search_author in auth_name.keys():
  print(auth_name[search_author])
else:
  print(" Oops.The author you are searching for is not found. Search another one")
```
